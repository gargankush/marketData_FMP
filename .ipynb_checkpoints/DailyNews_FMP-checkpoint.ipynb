# Daily stock news pulled from Financial Modelling Prep

### Important libraries required

```python
import requests
import pandas as pd
import csv
```

## Get the list of sysmbols for which financial statements are required

```python
ticker_data = pd.read_csv("S&P500Tickers.csv") # Open csv file and read data in pandas dataframe
companies = ticker_data['Symbol'].tolist() #Get all the tickers in a list
```

### Pulling the data from Financial Modeling Prep API and saving it in a dictionary

```python
company_news = []  # Declaring empty list
#final_news_df = pd.DataFrame() # Declearing final data frame
#news_data = pd.DataFrame() # Declearing intermediate data frame
companies = ['AMZN']
for company in companies:
    try:
        news = requests.get(f'https://financialmodelingprep.com/api/v3/stock_news?tickers={company}&limit=50&apikey=46e34bd7d2168d30561e3009d223dde3').json()
        #news_data = pd.DataFrame(news, columns=['symbol', 'publishedDate', 'title']) # Creating pandas data frame from list of dictionaries
        #final_news_df = final_news_df.append(news_data) # Appending to dataframe that will contain news for all the tickers
        #company_news.extend(news)
        company_news.append(news)
    except:
        pass
```

### Converting list to dataframe and writing that to csv file

```python
print(news)
```

```python
#news_df = pd.DataFrame(company_news, columns = ['symbol', 'publishedDate', 'title', 'image', 'site', 'text', 'url'])
# The array created in above step is list of list of distionaries. Therefore we will construct a list of data frames and then concatenate them
news_df = pd.concat([pd.DataFrame(d) for d in company_news])
#news_df = pd.DataFrame(company_news)
#news_df = pd.DataFrame.from_dict(company_news)
#pd.DataFrame.from_dict
#news_df = pd.DataFrame.from_dict(company_news)
#news_df.dropna(inplace=True)
news_df.to_csv('DailyNews_FMP.csv', mode='w', encoding = 'utf-8', index=False)
```

```python
file_name = 'DailyNews_FMP.csv'
file = open(file_name, 'w+', newline ='') # opening the csv file in 'w+' mode 
# writing the data into the file 
with file:     
    write = csv.writer(file) 
    write.writerows(company_news) 
```
